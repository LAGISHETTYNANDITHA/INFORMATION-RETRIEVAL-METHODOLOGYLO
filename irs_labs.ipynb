```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
sample_text ="The AI-Powered Code Review Assistant(AI-CRA) is an advanced tool designed to enhance the code review process by integrating Artificial intelligence with traditional review methods.It automates the detection of code issues, suggests improvements and ensures adherence to coding standards.By levering Machine learning and Natural Language preprocessing, AI -CRA helps developers write cleaner, more collaborative review process."
lower_text = sample_text.lower()

text_no_punct = lower_text.translate(str.maketrans('', '', string.punctuation))

words = word_tokenize(text_no_punct)

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

print("Original Text:\n", sample_text)
print("\nLowercased Text:\n", lower_text)
print("\nText without Punctuation:\n", text_no_punct)
print("\nTokenized Words:\n", words)
print("\nFiltered Words (without Stopwords):\n", filtered_words)
```

Perform Stemming, Tokenization, and Lemmatization

Objective: Explore various text normalization techniques.
Tasks:
•    Tokenize a sample text into words.
•    Apply stemming using algorithms like Porter or Snowball Stemmer.
•    Perform lemmatization using WordNetLemmatizer.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download required resources
nltk.download('punkt')
nltk.download('wordnet')

# Sample text
sample_text = "The children are playing in the playground. Running, jumped, and played are fun activities."

# Tokenization
words = word_tokenize(sample_text)

# Stemming using PorterStemmer
porter_stemmer = PorterStemmer()
stemmed_words_porter = [porter_stemmer.stem(word) for word in words]

# Stemming using SnowballStemmer
snowball_stemmer = SnowballStemmer("english")
stemmed_words_snowball = [snowball_stemmer.stem(word) for word in words]

# Lemmatization using WordNetLemmatizer
nltk.download('omw-1.4')  # Required for WordNet
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, wordnet.VERB) for word in words]  # Using VERB as the default POS

# Print results
print("Original Words:\n", words)
print("\nPorter Stemmer Output:\n", stemmed_words_porter)
print("\nSnowball Stemmer Output:\n", stemmed_words_snowball)
print("\nLemmatization Output:\n", lemmatized_words)
```

Perform Feature Extraction – Bag-of-Words

Objective: Learn how to represent text as numerical features using the Bag-of-Words model.

Tasks:

•    Prepare a corpus of sample text documents.
•    Convert text into a Bag-of-Words representation using NLTK or scikit-learn.
•    Visualize and analyze the term-document matrix.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus (collection of documents)
corpus = [
    "The cat sat on the mat.",
    "The dog barked at the mailman.",
    "Dogs and cats are common pets.",
    "The mailman was afraid of the barking dog."
]

# Convert text to Bag-of-Words representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

# Convert the matrix to a Pandas DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Print term-document matrix
print("Bag-of-Words Term-Document Matrix:\n")
print(bow_df)

# Visualize word frequencies
word_counts = np.asarray(X.sum(axis=0)).flatten()
words = vectorizer.get_feature_names_out()

# Create a bar plot
plt.figure(figsize=(10, 5))
plt.bar(words, word_counts, color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency in Corpus")
plt.xticks(rotation=45)
plt.show()
```

Implement Word Analysis and Word Generation

Objective: Conduct analysis and generation of words using Python libraries.
Tasks:
       •    Analyze word frequency and distributions.
       •    Use Markov models or GPT-based libraries to generate text.

```python
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Sample text corpus
text = """Machine learning is a subset of artificial intelligence. Artificial intelligence enables machines to learn.
          Learning algorithms power various AI applications. AI and ML are transforming the world."""

# Tokenization
nltk.download('punkt')
words = word_tokenize(text.lower())  # Convert to lowercase

# Frequency Distribution
fdist = FreqDist(words)

# Print most common words
print("Most Common Words:\n", fdist.most_common(5))

# Plot Word Distribution
plt.figure(figsize=(10, 5))
fdist.plot(15, title="Word Frequency Distribution")  # Plot top 15 words
plt.show()
```

Perform Morphological Analysis

Objective: Understand the structure of words and perform morphological parsing.

Tasks:
       •  Use NLTK’s morphological analyzer or other tools to parse words into root forms, prefixes, and suffixes.
       •  Perform exercises to analyze the morphological structure of given words.

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary resources
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample words for analysis
words = ["running", "happily", "nationalization", "cats", "unhappiness"]

# Stemming (reduces words to root form)
stemmer = PorterStemmer()
stemmed_words = {word: stemmer.stem(word) for word in words}

# Lemmatization (extracts base form)
lemmatizer = WordNetLemmatizer()
lemmatized_words = {word: lemmatizer.lemmatize(word, pos='v') for word in words}  # Using 'v' for verbs

# Print results
print("Stemming Results:", stemmed_words)
print("Lemmatization Results:", lemmatized_words)
```

```python
import re

# Function to extract prefixes and suffixes
def analyze_morphology(word):
    prefix_match = re.match(r"^(un|pre|re|mis|dis|in|im|non|over|under|sub|super|anti|auto)", word)
    suffix_match = re.search(r"(ing|ly|ed|es|er|ion|tion|s|ment|ness|ful|less|able|ive|ous|ism|ist|ize|al)$", word)

    prefix = prefix_match.group(0) if prefix_match else None
    suffix = suffix_match.group(0) if suffix_match else None

    return {"word": word, "prefix": prefix, "suffix": suffix}

# Sample words
words = ["unhappy", "running", "nationalization", "misinterpretation", "joyful", "carelessness"]

# Analyze words
morph_analysis = [analyze_morphology(word) for word in words]

# Print results
for analysis in morph_analysis:
    print(analysis)
```
